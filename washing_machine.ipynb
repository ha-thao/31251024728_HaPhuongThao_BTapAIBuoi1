```python
load = ctrl.Antecedent(np.arange(0, 11, 1), 'load')
dirt_level = ctrl.Antecedent(np.arange(0, 11, 1), 'dirt_level')
fabric_type = ctrl.Antecedent(np.arange(0, 11, 1), 'fabric_type')
dirt_type = ctrl.Antecedent(np.arange(0, 11, 1), 'dirt_type')


wash_time = ctrl.Consequent(np.arange(0, 121, 1), 'wash_time')
water_level = ctrl.Consequent(np.arange(0, 81, 1), 'water_level')
water_temp = ctrl.Consequent(np.arange(20, 91, 1), 'water_temp')




load['small'] = fuzz.trimf(load.universe, [0, 0, 5])
load['medium'] = fuzz.trimf(load.universe, [0, 5, 10])
load['large'] = fuzz.trimf(load.universe, [5, 10, 10])


dirt_level['low'] = fuzz.trimf(dirt_level.universe, [0, 0, 5])
dirt_level['normal'] = fuzz.trimf(dirt_level.universe, [0, 5, 10])
dirt_level['high'] = fuzz.trimf(dirt_level.universe, [5, 10, 10])


dirt_type['mo_hoi'] = fuzz.trimf(dirt_type.universe, [0, 0, 5])
dirt_type['bun_dat'] = fuzz.trimf(dirt_type.universe, [0, 5, 10])
dirt_type['dau_mo'] = fuzz.trimf(dirt_type.universe, [5, 10, 10])


fabric_type['delicate'] = fuzz.trimf(fabric_type.universe, [0, 0, 3])
fabric_type['regular'] = fuzz.trimf(fabric_type.universe, [2, 5, 8])
fabric_type['heavy'] = fuzz.trimf(fabric_type.universe, [7, 10, 10])


wash_time['short'] = fuzz.trimf(wash_time.universe, [0, 0, 25])
wash_time['medium'] = fuzz.trimf(wash_time.universe, [20, 35, 50])
wash_time['long'] = fuzz.trimf(wash_time.universe, [45, 70, 70])


water_level['low'] = fuzz.trimf(water_level.universe, [0, 0, 40])
water_level['medium'] = fuzz.trimf(water_level.universe, [0, 40, 80])
water_level['high'] = fuzz.trimf(water_level.universe, [40, 80, 80])


water_temp['lanh'] = fuzz.trimf(water_temp.universe, [20, 20, 55])
water_temp['am'] = fuzz.trimf(water_temp.universe, [20, 55, 90])
water_temp['nong'] = fuzz.trimf(water_temp.universe, [55, 90, 90])



# Tải Nhỏ
rule1 = ctrl.Rule(load['small'] & dirt_level['low'], [wash_time['short'], water_level['low']])
rule2 = ctrl.Rule(load['small'] & dirt_level['normal'], [wash_time['medium'], water_level['medium']])
rule3 = ctrl.Rule(load['small'] & dirt_level['high'], [wash_time['long'], water_level['medium']])

# Tải Trung Bình
rule4 = ctrl.Rule(load['medium'] & dirt_level['low'], [wash_time['medium'], water_level['medium']])
rule5 = ctrl.Rule(load['medium'] & dirt_level['normal'], [wash_time['long'], water_level['medium']])
rule6 = ctrl.Rule(load['medium'] & dirt_level['high'], [wash_time['long'], water_level['high']])

# Tải Lớn
rule7 = ctrl.Rule(load['large'] & dirt_level['low'], [wash_time['medium'], water_level['high']])
rule8 = ctrl.Rule(load['large'] & dirt_level['normal'], [wash_time['long'], water_level['high']])
rule9 = ctrl.Rule(load['large'] & dirt_level['high'], [wash_time['long'], water_level['high']])


# Vải Mỏng (Delicate)
rule10 = ctrl.Rule(fabric_type['delicate'] & dirt_type['mo_hoi'], [water_temp['lanh'], wash_time['short']])
rule11 = ctrl.Rule(fabric_type['delicate'] & dirt_type['bun_dat'], [water_temp['am'], wash_time['medium']])
rule12 = ctrl.Rule(fabric_type['delicate'] & dirt_type['dau_mo'], [water_temp['am'], wash_time['medium']])

# Vải Thường (Regular)
rule13 = ctrl.Rule(fabric_type['regular'] & dirt_type['mo_hoi'], [water_temp['lanh'], wash_time['medium']])
rule14 = ctrl.Rule(fabric_type['regular'] & dirt_type['bun_dat'], [water_temp['am'], wash_time['medium']])
rule15 = ctrl.Rule(fabric_type['regular'] & dirt_type['dau_mo'], [water_temp['nong'], wash_time['long']])

# Vải Dày (Heavy)
rule16 = ctrl.Rule(fabric_type['heavy'] & dirt_type['mo_hoi'], [water_temp['am'], wash_time['medium']])
rule17 = ctrl.Rule(fabric_type['heavy'] & dirt_type['bun_dat'], [water_temp['nong'], wash_time['long']])
rule18 = ctrl.Rule(fabric_type['heavy'] & dirt_type['dau_mo'], [water_temp['nong'], wash_time['long']])


# Mức độ bẩn Thấp (Low)
rule19 = ctrl.Rule(dirt_level['low'] & dirt_type['mo_hoi'], [wash_time['short'], water_temp['lanh']])
rule20 = ctrl.Rule(dirt_level['low'] & dirt_type['bun_dat'], [wash_time['short'], water_temp['lanh']])
rule21 = ctrl.Rule(dirt_level['low'] & dirt_type['dau_mo'], [wash_time['medium'], water_temp['am']])

# Mức độ bẩn Trung bình (Normal)
rule22 = ctrl.Rule(dirt_level['normal'] & dirt_type['mo_hoi'], [wash_time['medium'], water_temp['lanh']])
rule23 = ctrl.Rule(dirt_level['normal'] & dirt_type['bun_dat'], [wash_time['medium'], water_temp['am']])
rule24 = ctrl.Rule(dirt_level['normal'] & dirt_type['dau_mo'], [wash_time['long'], water_temp['nong']])

# Mức độ bẩn Cao (High)
rule25 = ctrl.Rule(dirt_level['high'] & dirt_type['mo_hoi'], [wash_time['medium'], water_temp['am']])
rule26 = ctrl.Rule(dirt_level['high'] & dirt_type['bun_dat'], [wash_time['long'], water_temp['nong']])
rule27 = ctrl.Rule(dirt_level['high'] & dirt_type['dau_mo'], [wash_time['long'], water_temp['nong']])

# Tải Nhỏ (Small)
rule28 = ctrl.Rule(load['small'] & fabric_type['delicate'], [water_level['low'], water_temp['lanh']])
rule29 = ctrl.Rule(load['small'] & fabric_type['regular'], [water_level['low'], water_temp['am']])
rule30 = ctrl.Rule(load['small'] & fabric_type['heavy'], [water_level['medium'], water_temp['am']])

# Tải Trung bình (Medium)
rule31 = ctrl.Rule(load['medium'] & fabric_type['delicate'], [water_level['medium'], water_temp['lanh']])
rule32 = ctrl.Rule(load['medium'] & fabric_type['regular'], [water_level['medium'], water_temp['am']])
rule33 = ctrl.Rule(load['medium'] & fabric_type['heavy'], [water_level['high'], water_temp['nong']])

# Tải Lớn (Large)
rule34 = ctrl.Rule(load['large'] & fabric_type['delicate'], [water_level['high'], water_temp['am']])
rule35 = ctrl.Rule(load['large'] & fabric_type['regular'], [water_level['high'], water_temp['am']])
rule36 = ctrl.Rule(load['large'] & fabric_type['heavy'], [water_level['high'], water_temp['nong']])

# Vải Mỏng (Delicate)
rule37 = ctrl.Rule(fabric_type['delicate'] & dirt_level['low'], [wash_time['short'], water_temp['lanh']])
rule38 = ctrl.Rule(fabric_type['delicate'] & dirt_level['normal'], [wash_time['medium'], water_temp['lanh']])
rule39 = ctrl.Rule(fabric_type['delicate'] & dirt_level['high'], [wash_time['medium'], water_temp['am']])

# Vải Thường (Regular)
rule40 = ctrl.Rule(fabric_type['regular'] & dirt_level['low'], [wash_time['short'], water_temp['lanh']])
rule41 = ctrl.Rule(fabric_type['regular'] & dirt_level['normal'], [wash_time['medium'], water_temp['am']])
rule42 = ctrl.Rule(fabric_type['regular'] & dirt_level['high'], [wash_time['long'], water_temp['nong']])

# Vải Dày (Heavy)
rule43 = ctrl.Rule(fabric_type['heavy'] & dirt_level['low'], [wash_time['medium'], water_temp['am']])
rule44 = ctrl.Rule(fabric_type['heavy'] & dirt_level['normal'], [wash_time['long'], water_temp['am']])
rule45 = ctrl.Rule(fabric_type['heavy'] & dirt_level['high'], [wash_time['long'], water_temp['nong']])

all_rules = [
    rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9,
    rule10, rule11, rule12, rule13, rule14, rule15, rule16, rule17, rule18,
    rule19, rule20, rule21, rule22, rule23, rule24, rule25, rule26, rule27,
    rule28, rule29, rule30, rule31, rule32, rule33, rule34, rule35, rule36,
    rule37, rule38, rule39, rule40, rule41, rule42, rule43, rule44, rule45
]

washing_ctrl = ctrl.ControlSystem(all_rules)
washing_sim = ctrl.ControlSystemSimulation(washing_ctrl)

washing_sim.input['load'] = 7.5
washing_sim.input['dirt_level'] = 8.0
washing_sim.input['fabric_type'] = 2.0
washing_sim.input['dirt_type'] = 9.0

try:
    washing_sim.compute()

    print("--- KẾT QUẢ ĐIỀU KHIỂN MÁY GIẶT ---")
    print(f"Thời gian giặt: {washing_sim.output['wash_time']:.2f} phút")
    print(f"Mức nước:      {washing_sim.output['water_level']:.2f} lít")
    print(f"Nhiệt độ nước: {washing_sim.output['water_temp']:.2f} °C")

except Exception as e:
    print(f"Lỗi tính toán: {e}")
    print("Kiểm tra xem tất cả các rule từ 1-45 đã được chạy (Run) chưa nhé!")
```
